# Final conversion

```python
h5_dir = '/Volumes/Elements/'
fn_zarr = '/Volumes/Elements_1/file1.zarr'
chunk = {'time':3000, 'distance':1721}
```

```python
def h52zarr_xarray(h5_dir, fn_zarr, df, chunk):
    # initial checkpoint
    checkpoint = 0
    first_loop=True
    for i in range(0, len(df)):
        current_file = h5_dir+df['File_Name'][i]
        hf = h5py.File(current_file)
        ds = xr.Dataset({
            'RawData':((['time', 'distance'], hf['Acquisition']['Raw[0]']['RawData'][:])),
            'RawDataSampleCount':('time', hf['Acquisition']['Raw[0]']['RawDataSampleCount'][:]),
            'RawDataTime':('time', hf['Acquisition']['Raw[0]']['RawDataTime'])})
        ds = ds.chunk(chunk)
        
        if first_loop==True:
            first_loop = False
            print('New zarr file created at index value', i)
            ds.to_zarr(fn_zarr, mode = 'w-')
            
        elif (df['hour'][i]==checkpoint):
            # if hour hits checkpoint, reset checkpoint
            checkpoint = 0 if df['hour'][i]==12 else 12 
            fn_zarr = '/Volumes/Elements_1/file'+str(i)+'.zarr' # new folder name
            print('New zarr file created at index value', i)
            ds.to_zarr(fn_zarr, mode = 'w-')
        else: 
            # else keep appending
            ds.to_zarr(fn_zarr, append_dim = 'time') 
    return 
```

```python
h52zarr_xarray(h5_dir, fn_zarr, df, chunk)
```
