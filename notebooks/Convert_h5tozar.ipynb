## Selecting files with same time dimension

```python
df = pd.read_csv('DAS_file_details.csv')
df.shape
```

```python
df = df.loc[df['Time_dimension']%3000==0]
df.reset_index(drop=True)
```

```python
df_sorted = df.sort_values(by='File_Name')
df_sorted.reset_index(drop=True)
```

```python
h5_files_path = '/Volumes/Elements/'
fn_zarr_path = '/Volumes/Elements_1/combined_timedim.zarr'
chunk = {'time':3000, 'distance':1721}
file_list = list(df_sorted['File_Name'])
```

```python
def h52zarr_xarray(h5_dir, fn_zarr, file_list, chunk):
#    file_list = os.listdir(h5_dir)
#     file_list.sort()
    files = [h5_dir+file for file in file_list]

    first_loop=True

    for file in tqdm(files):
        hf = h5py.File(file)
        ds = xr.Dataset({
            'RawData':((['time', 'distance'], hf['Acquisition']['Raw[0]']['RawData'][:])),
            'RawDataSampleCount':('time', hf['Acquisition']['Raw[0]']['RawDataSampleCount'][:]),
            'RawDataTime':('time', hf['Acquisition']['Raw[0]']['RawDataTime'])})

        ds = ds.chunk(chunk)
                
        # create new zarr store if beginning of loop otherwize, append in time dimension
        if first_loop:
            first_loop = False
            ds.to_zarr(fn_zarr, mode='w-')
        else:
            ds.to_zarr(fn_zarr, append_dim='time')
    return 
```

```python
h52zarr_xarray(h5_dir = h5_files_path, fn_zarr = fn_zarr_path, file_list = file_list, chunk = chunk)
```
